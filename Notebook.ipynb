### Manejo de valores NaN por columna
A continuación, se procederá con el manejo de valores NaN. Este apartado presenta casos muy particulares, ya que en algunas columnas la cantidad de datos NaN representa casi el 50% del total. Se abordará cada caso de manera individual, ya que la estrategia para manejar los valores NaN puede variar según la columna.

```python
# Primero se arreglarán los valores NaN de la columna "name" y "genre"

df_games.dropna( inplace = True, subset = ["name"] )
```

Comenzaremos con la columna `name` y en este caso al solucionar los valores NaN en esta columna también lo haremos en la columna `genre`. Se eliminaron los registros con valores NaN, ya que al tratarse de solo 2 registros, su impacto en el análisis es mínimo. Afortunadamente, tanto la columna `name` como la columna `genre` tenían sus valores nulos en los mismos registros. Por lo tanto, al eliminar los registros NaN de una columna, también se eliminaron de la otra, ahora podemos proceder con la siguiente columna.

```python
# Columna "year_of_release"

df_games.year_of_release.fillna( df_games.year_of_release.median(), inplace = True )
```

Para la columna `year_of_release`, se tomará otro enfoque. Dado que la cantidad de datos NaN en esta columna no es muy grande, pero tampoco es lo suficientemente pequeña como para eliminar los registros sin afectar el análisis, se procederá a reemplazar los valores NaN con la mediana de los datos. Esta metodología asegura que los datos faltantes se imputen de manera que refleje de forma más precisa la distribución general de los valores en la columna.

```python
# Columna "critic_score"

df_games.critic_score.fillna( 101, inplace = True )
```

La exploración de datos mostró que la cantidad de registros NaN en la columna `critic_score` es cercana al 50% del total de registros. Debido a esta alta proporción de datos faltantes, rellenar los valores NaN con la media o mediana no es viable, ya que esto podría introducir un sesgo significativo en los datos. Tampoco es conveniente eliminar los registros, ya que se perdería cerca del 50% de los datos.

Por lo tanto, la opción más adecuada es ignorar los registros con valores NaN al realizar cálculos, por varias razones:

- La principal es que la mayoría de los datos faltantes pertenecen a registros de videojuegos anteriores al 2010. Como veremos más adelante en el análisis de datos, los registros anteriores a 2010 no serán precisamente los protagonistas del análisis.
- Otra razón es que un estadístico calculado a partir de una muestra es aplicable al resto de la población.

Aunque se indicará que los valores NaN serán ignorados, esto no significa que se dejarán sin más. En su lugar, se reemplazarán con el valor "101". La intención de reemplazar los valores NaN con "101" es facilitar la conversión de esta columna a tipo entero en la siguiente etapa. Además, será más fácil filtrar los datos para ignorarlos de esta manera. Dado que el número "101" está fuera del rango de valores válidos para la columna (la columna `critic_score` solo contiene registros de 0 a 100), será fácilmente identificable.

```python
#En este segmento se demuestra que la cantidad de registros que faltan pertenecen a fechas anteriores al 2010
df = df_games.query( 'critic_score == 101' ).groupby( by='year_of_release', as_index=False ).count()

registers = df.query( 'year_of_release < 2010' ).critic_score.sum()

print( f"El número de datos NaN en la columna critic_score de antes del 2010 es de: {registers}")

registers = df.query( 'year_of_release >= 2010' ).critic_score.sum()

print( f"El número de datos NaN en la columna critic_score después del 2010 es de: {registers}")
```

El segmento de código anterior solo es para demostrar lo dicho anteriormente, la mayor cantidad de datos NaN pertenecen a juegos lanzados antes del 2010.

```python
# Columna "user_score"

df_games.user_score.fillna( 11, inplace = True )

df_games.user_score.replace( "tbd", 11 , inplace = True)
```

El caso de la columna `user_score` es casi idéntico al de la columna `critic_score`, por lo que la solución y las razones son las mismas, con dos pequeñas diferencias:

1. En lugar de reemplazar los valores NaN con "101", se reemplazarán con "11". Esto se debe a que la columna `user_score` almacena valores numéricos del 0 al 10, por lo que "11" es más coherente que "101".
2. En esta columna también se encuentran registros con el valor "tbd", que indica que aún no se ha asignado una calificación a esos juegos. Estos registros se tratarán de la misma manera que los valores NaN, es decir, se reemplazarán con "11". La razón para esto es que tanto NaN como "tbd" significan que el valor real es desconocido en ese registro, y por lo tanto, no se pueden considerar para el cálculo de estadísticas.

```python
#En este segmento se demuestra que la cantidad de registros que faltan pertenecen a fechas anteriores al 2010
df = df_games.query( 'user_score == 11' ).groupby( by='year_of_release', as_index=False ).count()

registers = df.query( 'year_of_release < 2010' ).user_score.sum()

print( f"El número de datos NaN en la columna user_score de antes del 2010 es de: {registers}")

registers = df.query( 'year_of_release >= 2010' ).user_score.sum()

print( f"El número de datos NaN en la columna user_score después del 2010 es de: {registers}")
```

De nuevo el bloque de código anterior solo demuestra que la mayoria de datos NaN pertenencen a juegos antiguos.

```python
# Columna "rating"

df_games.rating.fillna( "unknown" , inplace = True)
```

La última columna con valores NaN por corregir es la columna `rating`, y la situación es muy similar a las columnas anteriores. De igual manera, las razones y soluciones son bastante parecidas. En este caso, los valores NaN se reemplazarán con "unknown", para facilitar la manipulación y el filtrado de datos.

```python
#En este segmento se demuestra que la cantidad de registros que faltan pertenecen a fechas anteriores al 2010
df = df_games.query( 'rating == "unknown"' ).groupby( by='year_of_release', as_index=False ).count()

registers = df.query( 'year_of_release < 2010' ).rating.sum()

print( f"El número de datos NaN en la columna rating de antes del 2010 es de: {registers}")

registers = df.query( 'year_of_release >= 2010' ).rating.sum()

print( f"El número de datos NaN en la columna rating después del 2010 es de: {registers}")
```

### Cantidad de Videojuegos por Año

Para comenzar con el análisis, es fundamental entender la cantidad de videojuegos que se han producido a lo largo del tiempo. Este conocimiento nos permitirá observar cómo ha evolucionado el mercado y analizar posibles tendencias que puedan influir en las decisiones futuras. Identificar los años con mayor o menor producción de videojuegos nos dará una perspectiva clara sobre el crecimiento o declive del sector, lo que puede estar relacionado con factores tecnológicos, económicos o de demanda del consumidor.

```python
# Juegos lanzados cada año
df_graph = df_games.groupby( by = 'year_of_release', as_index = False ).count() # Agrupamos por año y aplicamos count() el metodo count

# Creación y configuración del gráfico de barras 
fig, ax = plt.subplots( figsize = (10,5) )
ax.bar( df_graph.year_of_release, df_graph.total_sales)
ax.set_title( "Reducción el volumen de producción anual.", fontsize = 25 )
ax.set_ylabel( "Cantidad de juegos lanzados.", fontsize = 14 )
ax.set_xlabel( "Años", fontsize = 14 )
ax.grid( True )

plt.show()
```

Desde entonces, la producción ha mantenido una estabilidad con poca variación año tras año.

- **Cambio de tendencia:** Tal vez la popularidad de los videojuegos esté experimentando un cambio, con nuevos tipos de entretenimiento ganando terreno.
- **Mayor complejidad:** La evolución tecnológica ha aumentado la complejidad de los videojuegos, lo que podría extender los tiempos de desarrollo. Los estudios pueden estar priorizando la calidad por encima de la cantidad, invirtiendo más tiempo en crear títulos más elaborados y atractivos.

Este cambio en la producción podría estar vinculado a factores internos del mercado o tendencias más amplias en el consumo y desarrollo de videojuegos.

### Selección de las plataformas más rentables

Ahora ya podemos proceder a seleccionar las consolas o plataformas en las que vamos a centrar nuestra atención. A primera vista, podría parecer que elegir las plataformas con mayores ventas sería la opción más adecuada. Sin embargo, en este mercado, las decisiones no son tan simples, ya que existen muchos otros factores que debemos considerar para una selección más precisa y estratégica.

```python
# Ventasde cada plataforma
df_graph = df_games.groupby( by = 'platform', as_index = False ).sum().sort_values( 'total_sales', ascending = False )

fig, ax = plt.subplots( figsize = (10,5) )

ax.bar( df_graph.platform, df_graph.total_sales)
ax.set_title( "Las plataformas con más ventas", fontsize = 25 )
ax.set_ylabel( "Ventas totales hasta 2016.", fontsize = 14 )
ax.set_xlabel( "Plataforma", fontsize = 14 )
plt.xticks( rotation = 90 )
ax.grid( True, )

```

El gráfico de barras anterior nos muestra las ventas totales de todas las plataformas, y se puede apreciar claramente cuáles son las que tienen más ventas. Si nos dejáramos guiar únicamente por esta gráfica, sin duda alguna diríamos que las plataformas a seleccionar deberían ser **PS2**, **X360**, **PS3**, **Wii**, **DS** y **PS**. Sin embargo, como veremos más adelante, tomar esta decisión sería un error significativo, ya que existen otros factores que deben ser considerados para identificar las plataformas más rentables.

```python
# Esta función recibe un conjunto de datos y un objeto axes de pyplot 
# y se encarga de crear graficas con un formato definido 
def build_graph( df, ax ):
    x = np.arange( len(df.index) )
    width = .5
    ax.bar( x , df, width )
    ax.set_title( df.name )
    ax.set_xticks( x )
    ax.set_xticklabels( df.index )
    ax.set_ylabel('Ventas en mdd')
    ax.set_xlabel('Año')
    ax.tick_params( 'x', rotation = 90 )
    ax.grid( True )

# Con ayuda de pivot_table() creamos un conjunto de datos que contendrá las ventas totales para cada año
# y cada plataforma
df_graph = df_games.pivot_table( 
    values = 'total_sales', 
    columns = 'year_of_release', 
    index = 'platform', 
    aggfunc = 'sum',
    fill_value = 0 )

#Creamos una nueva columna en el dataframe que contiene la suma de las ventas de todos los años.
df_graph[ 'total_sales' ] = df_graph.sum( axis = 1 )
# Ordenamos dataframe de la plataforma con más ventas a la plataforma con menos ventas.
df_graph.sort_values( by = 'total_sales', ascending = False, inplace = True )


# Creamos una matriz de gráficos de 4 columnas y 5 filas para gráficar las 20 plataformas más vendidas
fig, axs = plt.subplots( 5,4, figsize = (25,37) )

cont = 0

# Graficamos las ventas de cada año para todas las plataformas
for ax_col in axs:
    for ax_row in ax_col:
        df = df_graph.iloc[cont, : -1]
        df = df[df > 0]
        build_graph( df, ax_row  )  
        cont += 1
        
plt.show()
```

### Analizando el efecto de las críticas en las ventas

Ahora que hemos identificado las plataformas con mayor potencial de ventas, es crucial analizar qué tipo de videojuegos tienen más posibilidades de éxito. Para ello, comenzaremos evaluando si deberíamos considerar más la opinión de los críticos profesionales o de los propios usuarios a la hora de promocionar los videojuegos.

```python
# Configuramos el dataframe para analizar
df = df_games.query( 'platform == "PS4"' )
df = df.query( 'critic_score <= 100' )

fig, ax = plt.subplots(figsize = (10,5))

#Graficamos el scatter plot
ax.scatter(df.critic_score, df.total_sales)
#Configuramos el gráfico
ax.set_title('La correlación existe pero es baja.', fontsize=25)
ax.set_xlabel('Puntuación de los criticos')
ax.set_ylabel('Ventas globales en millones de dolares')

# Calculamos la correlación
correlation = df.critic_score.corr(df.total_sales)

plt.show()
print("############################################################################################")
print(f"#La correlación entre las ventas y las opiniones de los criticos es de: {correlation}#")
print("############################################################################################")
```

```python
df = df_games.query( 'platform == "PS4"' )
df = df.query( 'user_score <= 10' )

fig, ax = plt.subplots(figsize = (10,5))

ax.scatter(df.user_score, df.total_sales)
ax.set_title('No hay una correlación.', fontsize=25)
ax.set_xlabel('Puntuación de los usuarios')
ax.set_ylabel('Ventas globales en millones de dolares')

correlation = df.user_score.corr(df.total_sales)

plt.show()
print("##############################################################################################")
print(f"#La correlación entre las ventas y las opiniones de los usuarios es de: {correlation}#")
print("##############################################################################################")
```

### Géneros más rentables

Otro aspecto crucial para identificar los videojuegos más rentables es analizar los géneros. Algunos géneros tienen un mayor potencial de generar ingresos que otros. En esta sección, determinaremos cuáles géneros deben recibir nuestra atención para maximizar las ventas y mejorar la rentabilidad.

```python

# Filtramos el dataframe original para obtener solo los datos de las plataformas en las que estamos trabajando
df_filtered = df_games[ df_games.platform.isin( good_platforms ) ]

# Obtenemos todos los generos disponibles
genres_list = df_filtered.genre.unique()

#Creamos una lista vacía para almacenar datframes
list_df = []

# Para cada genero se obtendrán las ventas totales
for genre in genres_list:
    df = df_games.query( 'genre == @genre' ).total_sales
    list_df.append( df )


fig, ax = plt.subplots( figsize = (10,5))

# Graficamos el diagrama de caja
bplot = ax.boxplot(list_df, 
                   patch_artist=True,
                   showfliers = False,
                   meanline = True,
                   medianprops = dict( color = 'black' ))

# Configuramos el diagrama de caja
ax.set_title('Los generos más vendidos', fontsize=25)
ax.set_ylabel('Ventas globales en millobnes de dolares')
ax.set_xticks( np.arange(len(genres_list)) + 1 )
ax.set_xticklabels(genres_list)
ax.tick_params( 'x', rotation = 90 )
ax.grid( True )

for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

En resumen, los géneros más rentables son **shooter**, **platform** y **sports**, mientras que los menos rentables son **adventure**, **strategy** y **puzzle**.

Es muy evidente que en Japón, los géneros **Role-Playing** y **Action** son los más jugados, con una diferencia significativa frente a los otros géneros. ## Hipótesis 

#### 1. Hipótesis para las Calificaciones de Usuarios por Plataforma

**Hipótesis nula (H0):** Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son iguales.  
**Hipótesis alternativa (H1):** Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.

**Umbral alfa:** 0.05

**Criterio para probar la hipótesis:** Realizaremos una prueba t para dos muestras independientes.

**Razón para el criterio:** La prueba t para dos muestras independientes es adecuada para comparar las medias de dos grupos distintos.

```python

xbox_one_scores = df_games.query( 'platform == "XOne" & user_score < 11' ).user_score
pc_scores = df_games.query( 'platform == "PC" & user_score < 11' ).user_score

alpha = 0.05

############################################################################################################
#Realizar la prueba de Levene
stat_levene, p_levene = stats.levene(xbox_one_scores, pc_scores)

print(f"Estadístico de Levene: {stat_levene}")
print(f"Valor p de Levene: {p_levene}")

#Determinar si las varianzas son iguales
if p_levene > alpha:
    print("No hay diferencias significativas en las varianzas. Usar equal_var=True.")
    equal_var = True
else:
    print("Las varianzas son significativamente diferentes. Usar equal_var=False.")
    equal_var = False

print('////////////////////////////////////////////////////////////////////////////////////////////')
############################################################################################################

# Realizar la prueba t para muestras independientes
t_stat_platforms, p_value_platforms = stats.ttest_ind(xbox_one_scores, pc_scores, equal_var=equal_var)

# Imprimir los resultados
print(f"Prueba t para Xbox One vs PC:")
print(f"Estadístico t: {t_stat_platforms}")
print(f"Valor p: {p_value_platforms}")

if p_value_platforms < alpha:
    print(f"Rechazamos la hipótesis nula: las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.")
else:
    print(f"No podemos rechazar la hipótesis nula: las calificaciones promedio de los usuarios para Xbox One y PC son iguales.")

```

La hipótesis nula, que afirmaba que las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son iguales, fue rechazada. Esto indica que existe una diferencia significativa en las calificaciones promedio de los usuarios entre las dos plataformas.
